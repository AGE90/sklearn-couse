```python
import pandas as pd
import numpy as np
import course.utils.paths as path


from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score, KFold
```

```python
data_DIR = path.data_raw_dir('happiness.csv')
dataset = pd.read_csv(data_DIR)
display(dataset)
```

```python
X = dataset.drop(['country', 'rank', 'score', 'high', 'low'], axis=1)
y = dataset['score']
```

```python
estimator = DecisionTreeRegressor()
score = cross_val_score(estimator, X, y, cv=4, scoring='neg_mean_squared_error')
print(score)
print('Score mean:', np.abs(np.mean(score)))
```

```python
kf = KFold(n_splits=4, shuffle=True, random_state=42)
for train, test in kf.split(dataset):
    print('Training set:', train)
    print('Test set:', test)
```
