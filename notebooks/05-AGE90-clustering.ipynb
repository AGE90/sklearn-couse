```python
import pandas as pd
import course.utils.paths as path

from sklearn.cluster import MiniBatchKMeans, MeanShift
```

```python
data_DIR = path.data_raw_dir('candy.csv')
dataset = pd.read_csv(data_DIR)
display(dataset)
```

```python
X = dataset.drop(['competitorname'], axis=1)
```

```python
kmeans = MiniBatchKMeans(
    n_clusters=4,
    batch_size=8,
    n_init='auto'
).fit(X)

print('Number of centroids:', len(kmeans.cluster_centers_))
print('-'*10)
dataset['labels_km'] = kmeans.labels_
display(dataset)
```

```python
mshift = MeanShift().fit(X)

print('Number of centroids:', len(mshift.cluster_centers_))
print('-'*10)
dataset['labels_ms'] = mshift.labels_
display(dataset)
```
